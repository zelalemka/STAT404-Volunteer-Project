```r
data = read.csv("305 Student Data.csv")
head(data)
attach(data)
```

```r
# Set Enrollment.time as a factor
data$Enrollment.time = factor(data$Enrollment.time)
levels(data$Enrollment.time)
```

```r
# Exclude the "Overall.grade" column from the list of predictors
predictors <- data[, !(names(data) %in% c("Overall.grade"))]

# Create individual scatterplots for each predictor against "Overall.grade"
for (col in colnames(predictors)) {
  plot(data[, col], data$Overall.grade, main = paste("Scatterplot of", col, "vs. Overall.grade"), xlab = col, ylab = "Overall.grade")
    
}
```

From the scatterplot, we can see that there is a correlation between the overall.grades and all the predictors. From initial observation, it seems that `Overall.grades` has a positive linear relationship with `enjoyment` and `Weekly.study hours`. `Overall.grades` has a negative linear relationship with `Class.absences`. We will verify this by calculating the correlation between the variables.


```r
# Exclude "Enrollment.time" because it is a factor
predictors_without_factor <- data[, !(names(data) %in% c("Overall.grade", "Enrollment.time"))]
correlations <- sapply(predictors_without_factor, function(col) cor(data$Overall.grade, col))
print(correlations)
```
